```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import data into data frame
df = pd.read_csv("injury_data_updated copy.csv")
df.head()
```

```python
#Does the amount of players out vs questionable change as the season progresses?

# Filter data for out and questionable
filtered_df = df[df["Injury Status"].isin(["Out", "Questionable"])]

# Group by Week and count the Injury Status
count_df = filtered_df.groupby(["Week", "Injury Status"]).size().unstack(fill_value=0)

#plot
count_df.plot(kind="bar", stacked=True)
plt.title("Injury Status Each Week", fontsize=20)
plt.xlabel("Week", fontsize=18)
plt.ylabel("Number of Players", fontsize=18)
plt.legend(title="Injury Status")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Group by Injury Status and count the number of players
count_df = filtered_df.groupby("Injury Status").size()

# Plot pie chart
count_df.plot(kind="pie", autopct='%1.1f%%', startangle=90)
plt.title("Percentage of Players Out vs Questionable",fontsize=22)  
plt.tight_layout()
plt.show()
```
